In this notebook, we will use numpy and matplotlib to detect edge of a image

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

Data loading and splitting colors channel to RGB

```python
img = Image.open('dataset_gambar.jpeg')
red,green,blue = img.split()
red = np.array(red)
green = np.array (green)
blue = np.array(blue)
```

Showing source image

```python
plt.imshow(img)
```

Check matrix shape of the image

```python
np.array(img).shape
```

```python
pixel_array = np.array([red, green, blue])
```

```python
grey = np.array(0.299 * red + 0.587 * green + 0.114 * blue)
```

The weights used in this conversion are calibrated for contemporary CRT phosphors: Y = 0.2125 R + 0.7154 G + 0.0721 B

```python
dx,dy = np.gradient(grey)
```

```python
mag = np.sqrt(np.power(dx,2)+np.power(dy,2))
```

An edge in the image represents a region where there is a sharp contrast, i.e. a rapid change of intensity. The main idea of the techniques for
edge detection is the computation of the gradient of the image.

Showing image that has been grey scaled

```python
plt.imshow(grey)
```

Showing image with gradient magnitude

```python
plt.imshow(np.where(mag>10,mag,0)) 
```

```python
plt.imshow(np.where(mag>30,mag,0))
```

```python
plt.imshow(np.where(mag>50,mag,0))
```

The edge detection method with magnitude > 10 was to wide. The background is still detected. Compared to magnitude > 30, the background is not detected, but the foreground still detected. And the gradient > 50, the object is hardly detected.
